# Assimetria

Conhecido como o valor que resume a não simetria do conjunto de dados em relação a média. Para interpretação dos valores, consideram-se as seguintes análises:

- Quanto mais próximo ao zero, mais simétrico é o conjunto de dados;
- Quando o valor estiver positivo, indica que há assimetria positiva do conjunto, ou seja, mais da metade dos valores se encontra acima da média;
- Quando o valor estiver negativo, indica que há assimetria negativa do conjunto, ou seja, mais da metade dos valores se encontra abaixo da média.


Lista de assimetria por coluna:

```python
dataset.skew()
```

Boxplot da coluna bedrooms, com estilo pré-definido:

```python
%matplotlib inline
matplotlib.style.use('tableau-colorblind10')
dataset.boxplot(column='bedrooms')
```

Boxplot da coluna price separado pela coluna bedrooms:

```python
%matplotlib inline
matplotlib.style.use('tableau-colorblind10')
dataset.boxplot(column='price', by='bedrooms')
```

# Correlação

A correlação é um valor que indica o quanto duas variáveis possuem semelhança quanto a variação juntas, seja positiva ou negativa. Em projetos de Machine Learning, saber quais as variáveis que possuem bons índices de correlação, facilitam na escolha das variáveis preditoras do modelo.

Os principais métodos utilizados para o cálculo de correlação são Pearson e Spearman. 

Correlação de momento de produto de Pearson: Esta avalia a relação linear entre duas variáveis contínuas. Uma relação é linear quando a mudança em uma variável é associada a uma mudança proporcional em outra variável. Como exemplo, poderia ser usada uma correlação de Pearson para avaliar se aumentos na temperatura de um forno de têmpera estão associados a uma redução da espessura de uma barra de ferro temperada.

Correlação da ordem de posto de Spearman: Esta avalia a relação monotônica entre duas variáveis contínuas ou ordinais. Em uma relação monotônica, as variáveis tendem a mudar juntas mas não necessariamente a uma taxa constante. O coeficiente de correlação de Spearman baseia-se nos valores classificados de cada variável, em vez de os dados brutos. Este método é usado para avaliar relações envolvendo variáveis ordinais. Por exemplo, poderia ser usada a correlação de Spearman para avaliar se a ordem na qual os funcionários executam um teste está relacionada ao número de meses de emprego.

Tabela de correlação entre colunas, com a função .corr(). Esta função traz automaticamente o cálculo de correlação considerando o método de Pearson:

```python
dataset.corr()
```

Tabela de correlação considerando o método de Spearman:

```python
dataset.corr('spearman')
```

Tabela considerando apenas colunas pré-determinadas, para facilitar a visualização:

```python
dataset[['bedrooms','bathrooms','sqft_living','floors','waterfront','grade','price']].corr()
```

Gráfico de linhas para visualização da correlação entre variáveis pré-definidas:

```python
%matplotlib inline
matplotlib.style.use('tableau-colorblind10')
dataset[['sqft_living','waterfront','grade','price']].corr().plot()
```

# Tabela Pivotada

Visando o agrupamento dos dados, a tabela pivotada simplifica a visualização dos dados, funciona como uma tabela dinâmica no excel.

Tabela pivotada com contagem de ID, segmentado por vista pro mar (0 = sem vista pro mar, 1 = com vista pro mar) e a quantidade de pisos na casa:

```python
dataset.pivot_table('id',index=["waterfront","floors"], aggfunc='count',margins=True)       
```

# Crosstab

Outra função que também pode ser usada é a Crosstab, onde o Pandas cruza os valores das variáveis e mostra a frequência com que aparecem no dataframe.

Crosstab da coluna bedrooms pela coluna condition:

```python
pd.crosstab(dataset['bedrooms'],dataset['condition'])
```

Gráfico de barras que apresenta a Crosstab supracitada de forma visual:

```python
%matplotlib inline
matplotlib.style.use('tableau-colorblind10')
table = pd.crosstab(dataset['bedrooms'],dataset['condition'])
table.plot(kind='bar',width=1.0, title='Condition by Bedrooms' ,grid=False)
```
