Importando o dataset load_iris do sklearn.datasets:

```python
from sklearn.datasets import load_iris

X = load_iris().data
y = load_iris().target
```

Criação de dataframe pandas a partir de X, y. Antes, concatenei os arrays numpy. Na sequência transformei o array final para um dataframe. Alterei o nome das colunas para 'X1', 'X2', 'X3', 'X4' e 'target'.

```python
iris_c_ = np.c_[X,y]
print(iris_c_)
print(iris_c_.shape)
```

```python
df_iris = pd.DataFrame(iris_c_, columns=["X1", "X2", "X3", "X4", "TARGET"])
```

Cálculo de média da variável 'X2', para cada valor de target com a função groupby:

```python
df_iris.groupby(by = "TARGET").apply(lambda x: x.X2.mean() )
```

```python
df_iris_map = df_iris.assign(TARGET_2=y)
df_iris_map.TARGET_2 = df_iris.TARGET.map({0:"A", 1:"B", 2:"C"})
```

Criação de uma função divide_dados:
        
        - recebe um array X de dimensões quaisquer (n,k) e um parâmetro p entre 0 e 1.
        - divide o array X em duas partes disjuntas de forma aleatória. Ou seja, crie dois novos arrays, X1 e X2,
        onde a dimensão de linhas de X1 somada com a dimensão de linhas de X2 deve ser igual a dimensão de linhas de X.
        - interpretando p como percentual, o tamanho de X1 deve ser p*100% do número de linhas de X; enquanto X2 deve ser o resto.
        - retorna esses dois arrays.

```python
x =  np.random.randint(1, 1000, size=(8, 2))
print(x)
print('------------------------*------------------------')
p = np.random.randint(1,7)
print(p)
print('------------------------*------------------------')
x1 = x[0:p]
print(x1)
print(type(x1))
print('------------------------*------------------------')
x2 = x[p:8]
print(x2)
print(type(x2))
```
