# União de dataframes

Considerando dois dataframes distintos, serão utilizadas algumas funções conforme consultas em linguagem SQL.

Leitura dos dados da tabela olist_orders_dataset em tabela csv:

```python
orders = pd.read_csv('datasets/olist_orders_dataset.csv')
```

Exibição do início do dataframe:

```python
orders.head()
```

Leitura dos dados da tabela olist_items_dataset em tabela csv:

```python
orders_items = pd.read_csv('datasets/olist_order_items_dataset.csv')
```

Exibição do início do dataframe:

```python
orders_items.head()
```

Criação de uma query com a função merge(), denifindo a coluna order_id como chave de ligação entre tabelas:

```python
query = pd.merge(orders[['order_id','order_status','order_approved_at']],
                 orders_items[['order_id','product_id','seller_id','price','freight_value']],
                 on='order_id'
)
```

Exibição da query realizada:

```python
query.head()
```

Criação de uma query com a função merge(), denifindo a coluna order_id como chave de ligação entre tabelas, consultando todos os pedidos independente da associação com itens (left join):

```python
query = pd.merge(orders[['order_id','order_status','order_approved_at']],
                 orders_items[['order_id','product_id','seller_id','price','freight_value']],
                 on='order_id',
                 how='left'
)
```

Contagem de valores nulos da query:

```python
query.isnull().sum()
```

Criação de uma query com a função merge(), denifindo a coluna order_id como chave de ligação entre tabelas, consultando todos os itens independente da associação com pedidos (right join):

```python
query = pd.merge(orders[['order_id','order_status','order_approved_at']],
                 orders_items[['order_id','product_id','seller_id','price','freight_value']],
                 on='order_id',
                 how='right'
)
```

Contagem de valores nulos da query:

```python
query.isnull().sum()
```

Criação de uma query com a função merge(), denifindo a coluna order_id como chave de ligação entre tabelas, consultando todos os itens (outer join):

```python
query = pd.merge(orders[['order_id','order_status','order_approved_at']],
                 orders_items[['order_id','product_id','seller_id','price','freight_value']],
                 on='order_id',
                 how='outer'
)
```

Contagem de valores nulos da query:

```python
query.isnull().sum()
```
