# Manipulações de arquivos Excel

Atualmente, grande parte dos dataframes a serem manipulados se encontram em excel, seja csv, xls ou xlsx, existem formas de transferir estes dados para o Pandas.

Importação da biblioteca Pandas para manipulação:

```python
import pandas as pd
```

Criação do dataframe usando a planilha em xlsx:

```python
dataframe_excel = pd.read_excel('Controle-de-Atividades-2.0.xlsx', header=1)
```

Criação do dataframe indexando a coluna Estado Atual:

```python
file = 'Controle-de-Atividades-2.0.xlsx'
dataframe_excel = pd.read_excel(file, header=1, index_col=3)
```
